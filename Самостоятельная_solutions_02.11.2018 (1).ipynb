```python
st = "Our man Jack is King of the Pumpkin patch Everyone hail to the Pumpkin King now"
lst = st.split(' ')
unique_words = []
for word in lst:
    if word not in unique_words:
        unique_words.append(word)
unique_words.reverse()
res = ' '.join(unique_words)
```

```python
tel_dict = {'Иванов Иван': ['+7 123 456-78-90'],
            'Петров Петр': ['+7 910 123-12-12', '+79161231231', '+7 (925) 123-12-12'],
            'Михайлов Михаил': ['+7 915 111 22', '+7(915999)-88-77'],
            'Романов Роман': ['+7945 123-45-67', '+7 949 000 11 22']}
```

```python
import re
```

```python
tel_pattern = re.compile(r"""^\+7
(  # случай 1: для первых четырех типов номеров
[ (][(]?
\d{3}
[ )][ ]?
\d{3}
([- ]?\d{2}){2}
)  # конец случай 1
|
(  # отдельный случай для пятого типа
\d{10}
)  # конец отдельный случай
$""", re.X)
```

```python
lst = []
for k, v in tel_dict.items():
    tels = []
    for tel in v:
        if tel_pattern.search(tel):
            tels.append(tel)
    st = '{} - {}'.format(k, ','.join(tels))
    lst.append(st)
res = '\n'.join(lst)
```

```python
input_dict = {'ноль': 0,
              'один': 1,
              'два': 2,
              'три': 3,
              'четыре': 4,
              'пять': 5,
              'шесть': 6,
              'семь': 7,
              'восемь': 8,
              'девять': 9,
              'десять': 10}
expr = input().split(' ')
operations = expr[1::2]
numb = []
for i in expr[::2]:
    numb.append(input_dict[i])
znach = numb[0]
for i, symb in enumerate(operations, 1):
    if symb == 'сложить':
        znach += numb[i]
    if symb == 'вычесть':
        znach -= numb[i]
    if symb == 'умножить':
        znach *= numb[i]
    if symb == 'разделить':
        znach /= numb[i]
znach
    
```

```python
tel_str = """Иванов Иван - +7 123 456-78-90
Петров Петр - +7 910 123-12-12,+79161231231,+7 (925) 123-12-12
Михайлов Михаил - +7 915 111 22,+7(915999)-88-77
Романов Роман - +7945 123-45-67,+7 949 000 11 22"""
```

```python
import re
```

```python
tel_pattern = re.compile(r"""^\+7
(  # случай 1: для первых четырех типов номеров
[ (][(]?
\d{3}
[ )][ ]?
\d{3}
([- ]?\d{2}){2}
)  # конец случай 1
|
(  # отдельный случай для пятого типа
\d{10}
)  # конец отдельный случай
$""", re.X)
```

```python
tel_dict = {}
for row in tel_str.split('\n'):
    name, tels = row.split(' - ')
    tels_lst = []
    for tel in tels.split(','):
        if tel_pattern.search(tel):
            tels_lst.append(tel)
    tel_dict[name] = tels_lst
```
