3) Реализовать функцию ``find_cathetus()`` , осуществляющую поиск второго катета по длинне гипотенузы и первого катета с проверкой входных значений и генерацией исключений с содержательными сообщениями при получении некорректных параметров.

````python
import math
````

````python
def find_cathetus(gip, kat):
    if gip <= 0 or kat <= 0:
        raise Exception('Такого трегольника не существует')
    if not isinstance(kat, (int, float)) or not isinstance(gip, (int, float)): #функция, которая проверяет Является ли этот демент инт и флоат
        raise Exception('Катет и гипотенуза должны быть числами')
    try:
        return math.sqrt(gip*gip - kat*kat)
    except ValueError:
        raise Exception('Треугольника с такими сторонами не существует')
        
````

5) Написать функцию ``write_file()``, которая просит пользователя построчно ввести содержимое текстового файла. Сначала запрашивается имя файла. После ввода пустой строки ввод заканчивается и файл сохраняется. Вызвать эту функцию, корректно обработать исключительные ситуации, возникающие при работе функции. 

````python
def write_file():
    filename = input('Введите имя файла: ')
    try:
        with open(filename, mode = 'w') as f:
            while True: 
                line = input('Введите строку для записи: ')
                if not line: #проверяем строку на пустоту
                    break
                f.write(line + '\n')
    except (OSError, ValueError, ZeroDivisionError):
        raise Exception('Имя файла содержит запрещённые символы')
    except Exception:
        print('Что-то пошло не так!!!')
````

6) Написать функцию ``print_file()``, которая построчно выводит на экран содержимое текстового файла. У функции должен быть необязательный параметр, при помощи которого можно задать максимальное количество выводимых на экран строк.

````python
def print_file(filename, m):
    x = 0
    with open(filename, mode = 'r') as f:
        for line in f:
            if x < m:
                print(line)
                x += 1
````

14) Создать новые файлы ``cinema2.json`` и ``cinema2.csv``, в которых будут записаны следующие поля из исходного файла ``cinema.json``:

* ID - поле ``global_id`` исходного файла;
* краткое наименование (поле ``CommonName``);
* адрес (текстовой строкой, только улица-дом, без индекса и района);
* сайт;
* рабочие часы по дням недели (в формате словаря "день недели: часы работы");
* долгота;
* широта.

````python
def generate_dict(json_file):
    def working_hours(works):
        dw = {}
        for i in works:
            dw[i['DayWeek']] = i['WorkHours']
        return dw

    def geo_data(geo):
        return geo['coordinates'][0][0], geo['coordinates'][0][1]

    lst = list()
    for i in json_file:
        d = {}
        d['id'] = i['global_id']
        d['Name'] = i['CommonName']
        d['Address'] = i['ObjectAddress'][0]['Address']
        d['DayWeek'] = working_hours(i['WorkingHours'])
        d['WebSite'] = i['WebSite']
        d['lat'],d['lon'] = geo_data(i['geoData'])
        lst.append(d)
    return lst
````

````python
d = generate_dict(json_file)
````

````python
with open(r'F:\Универ\Алгоритмы и структуры данных на языке Python\sem10.papka\cinema2.csv', 'w') as csv_file:
    headers = d[0].keys()
    csv_writer = csv.DictWriter(csv_file, fieldnames = headers)
    csv_writer.writeheader()
    csv_writer.writerows(d)
````

````python
with open(r'F:\Универ\Алгоритмы и структуры данных на языке Python\sem10.papka\cinema2.json', 'w', encoding='cp1251') as j_f:
    j_f.write(json.dumps(d, ensure_ascii=False))
````
